```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.plot(x='Country', y='Sales', kind='line')
plt.title("Line Plot")
plt.xlabel("Category")
plt.ylabel("Values")
plt.show()
```

```python
df.set_index('Profit')['Quantity'].plot(kind='pie', autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

```python
df['Profit'].plot(kind='hist', bins=5)
plt.title("Histogram")
plt.xlabel("Profit")
plt.show()
```
